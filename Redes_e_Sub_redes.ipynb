##**Introdução Teórica**

A arquitetura da Internet baseia-se no protocolo IP, que utiliza endereçamentos lógicos para identificar dispositivos. Segundo Nunes (2017), a divisão de uma rede principal em **Sub-redes** é uma estratégia essencial para otimizar o tráfego, melhorar a segurança e facilitar a gerência administrativa. Através do uso de **Máscaras de Sub-rede**, o Sistema Operacional consegue distinguir qual porção do endereço IP identifica a rede e qual identifica o host (dispositivo). O entendimento da notação **CIDR** (*Classless Inter-Domain Routing*) é fundamental para o profissional de ADS, pois permite a alocação flexível de endereços sem as limitações das antigas classes A, B e C.

###**1. Diagnóstico de Interface: IP e Máscara**

Para que um host se comunique, ele precisa saber a qual sub-rede pertence. Vamos extrair essas informações do ambiente atual.

```python
# Verificando as configurações de rede da máquina virtual (Interface eth0)
!ip addr show eth0 | grep inet
```

####**O que observar:**
<a>O output exibe o endereço IP seguido da máscara em notação CIDR (ex: /24). Como explica Nunes, essa barra indica quantos bits são dedicados à rede. Se a máscara for /24, os primeiros 24 bits (três primeiros octetos) são imutáveis para todos os dispositivos daquela mesma sub-rede.</a>

###**2. Simulação de Cálculo de Rede e Broadcast**

Todo segmento de rede possui dois endereços reservados: o endereço da Rede (primeiro) e o de Broadcast (último). Vamos simular o cálculo que o SO faz internamente.

```python
import ipaddress

# Definindo uma rede IP com base na teoria da Seção 3.1 (Exemplo: 192.168.10.0/26)
# Uma máscara /26 divide uma classe C em 4 sub-redes menores
rede_exemplo = ipaddress.ip_network('192.168.10.0/26')

print(f"Rede Definida: {rede_exemplo}")
print(f"Mascara de Sub-rede: {rede_exemplo.netmask}")
print(f"Endereco de Rede: {rede_exemplo.network_address}")
print(f"Endereco de Broadcast: {rede_exemplo.broadcast_address}")
print(f"Intervalo de Hosts Validos: {rede_exemplo[1]} ate {rede_exemplo[-2]}")
```

####**O que observar:**
<a>Neste teste, simulamos uma sub-rede menor (/26). Note que o endereço de broadcast é o último do intervalo. De acordo com o livro, o broadcast é utilizado para enviar dados a todos os hosts daquela sub-rede simultaneamente, sem interferir em outras sub-redes vizinhas.</a>

###**3. Lógica de Roteamento: Segmentação e Gateway**

Nesta etapa, simularemos o processo de decisão da Camada de Rede. O Sistema Operacional utiliza a máscara de sub-rede para determinar se o destino é local ou se o pacote deve ser encaminhado para o roteador (Gateway), garantindo a segmentação lógica descrita por Nunes.

```python
import ipaddress

# 1. Definindo nossa configuração local (Baseada no Teste 1)
minha_rede = ipaddress.ip_network('192.168.1.0/24')
meu_ip = ipaddress.ip_address('192.168.1.15')

# 2. Definindo dois destinos para teste
destino_local = ipaddress.ip_address('192.168.1.50')
destino_remoto = ipaddress.ip_address('8.8.8.8')

def validar_rota(ip_destino):
    print(f"[*] Analisando destino: {ip_destino}")
    if ip_destino in minha_rede:
        print(f"[LOGICA] O destino pertence a rede {minha_rede}. Entrega DIRETA via Camada 2.")
    else:
        print(f"[LOGICA] O destino FORA da rede local. Encaminhando para o GATEWAY PADRAO.")

# Executando a validação
validar_rota(destino_local)
print("-" * 30)
validar_rota(destino_remoto)
```

####**O que observar:**
<a>O teste demonstra a função prática da máscara de sub-rede explicada na Seção 3.1 do livro de Nunes.
>**Nota Técnica**: Devido às restrições de segurança do ambiente Google Colab, que impedem o tráfego de pacotes ICMP (Ping), utilizamos uma simulação lógica da pilha TCP/IP. O experimento prova que o Sistema Operacional realiza uma operação AND entre o IP de destino e a máscara local; se o resultado for externo à rede, o pacote é obrigatoriamente enviado ao Gateway, validando o conceito de roteamento e segmentação.</a>
